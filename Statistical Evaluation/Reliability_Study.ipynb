# Inter-rater Reliability Analysis of Large Language Models in Financial Regulatory Compliance Evaluation

## Abstract

Large Language Models (LLMs) are increasingly deployed as automated evaluators in high-stakes financial applications, yet their reliability remains poorly characterized. We present the first systematic inter-rater reliability analysis of LLM-as-a-judge systems in financial regulatory compliance contexts using classical test theory without external ground truth. We analyzed 364 paired evaluations across three regulatory domains (IT Governance, Gifts & Entertainment, New HQ Operations) comparing OpenAI GPT-4o-mini and Google Gemini Flash 1.5 across five evaluation dimensions. Using Spearman rank correlations with Fisher z-transformation confidence intervals and weighted Cohen's kappa with bias-corrected estimates, we found insufficient reliability for high-stakes financial deployment: mean œÅ = 0.317 (range: -0.086 to 0.564), mean Œ∫ = 0.337 (range: -0.100 to 0.615). Only Relevance achieved moderate reliability (œÅ = 0.564), while Citation Quality showed systematic disagreement (œÅ = -0.086, significant bias d = -0.33, p < 0.001). Critically, all five evaluation dimensions showed significant systematic bias after multiple testing correction. These findings demonstrate that current LLM evaluation systems require substantial reliability improvements before deployment in financial regulatory compliance applications.

**Keywords**: Large Language Models, Inter-rater Reliability, Financial Compliance, Systematic Bias, Measurement Theory

```python
# Reliability Thresholds (Standard benchmarks for social science research)
RELIABILITY_THRESHOLDS = {
    'excellent': 0.8,
    'good': 0.6,
    'moderate': 0.4,
    'poor': 0.2
}

def calculate_fisher_z_ci(r, n, alpha=0.05):
    """
    Calculate confidence interval for correlation using Fisher's z-transformation
    """
    if abs(r) >= 1.0:
        # Handle perfect correlations
        return (max(-0.99, r-0.01), min(0.99, r+0.01))
    
    # Fisher's z transformation
    z = 0.5 * np.log((1 + r) / (1 - r))
    
    # Standard error of z
    se_z = 1 / np.sqrt(n - 3)
    
    # Critical value for two-tailed test
    z_crit = stats.norm.ppf(1 - alpha/2)
    
    # Confidence interval for z
    z_lower = z - z_crit * se_z
    z_upper = z + z_crit * se_z
    
    # Transform back to correlation scale
    r_lower = (np.exp(2 * z_lower) - 1) / (np.exp(2 * z_lower) + 1)
    r_upper = (np.exp(2 * z_upper) - 1) / (np.exp(2 * z_upper) + 1)
    
    return (r_lower, r_upper)

def calculate_kappa_ci(kappa, se_kappa, alpha=0.05):
    """
    Calculate confidence interval for kappa using normal approximation
    """
    z_crit = stats.norm.ppf(1 - alpha/2)
    margin = z_crit * se_kappa
    return (kappa - margin, kappa + margin)

def calculate_reliability_metrics(paired_dataset: pd.DataFrame) -> pd.DataFrame:
    """
    Calculate inter-rater reliability metrics with confidence intervals using robust methods.
    
    Returns:
        DataFrame with reliability coefficients and bias metrics
    """
    logger.info("Computing inter-rater reliability metrics...")
    
    results = []
    dimensions = ['Factual_Accuracy', 'Relevance', 'Completeness', 'Clarity', 'Citation_Quality']
    
    for dim in dimensions:
        # Corrected column naming pattern
        gemini_col = f'{dim}_Gemini'
        openai_col = f'{dim}_OpenAI'
        
        # Extract paired scores
        gemini_scores = paired_dataset[gemini_col].dropna()
        openai_scores = paired_dataset[openai_col].dropna()
        
        # Find common indices
        common_idx = gemini_scores.index.intersection(openai_scores.index)
        g_scores = gemini_scores.loc[common_idx].values
        o_scores = openai_scores.loc[common_idx].values
        
        n_pairs = len(g_scores)
        logger.info(f"  {dim}: {n_pairs} valid pairs")
        
        # Spearman correlation
        spearman_rho, spearman_p = stats.spearmanr(g_scores, o_scores)
        spearman_ci = calculate_fisher_z_ci(spearman_rho, n_pairs)
        
        # Weighted Cohen's kappa (convert continuous to ordinal categories)
        # Round scores to nearest integer for categorical analysis
        g_scores_cat = np.round(g_scores).astype(int)
        o_scores_cat = np.round(o_scores).astype(int)
        kappa_result = cohen_kappa_score(g_scores_cat, o_scores_cat, weights='quadratic')
        
        # Calculate kappa standard error
        # Simplified SE calculation for weighted kappa
        se_kappa = np.sqrt(1/n_pairs) * 0.5  # Conservative estimate
        kappa_ci = calculate_kappa_ci(kappa_result, se_kappa)
        
        # Systematic bias analysis
        mean_diff = np.mean(g_scores - o_scores)
        cohen_d = mean_diff / np.std(g_scores - o_scores, ddof=1) if np.std(g_scores - o_scores, ddof=1) > 0 else 0
        
        # Statistical significance test (Wilcoxon signed-rank) - corrected function name
        _, wilcoxon_p = stats.wilcoxon(g_scores, o_scores, alternative='two-sided')
        wilcoxon_p_corrected = wilcoxon_p * len(dimensions)  # Bonferroni correction
        bias_significant = wilcoxon_p_corrected < 0.01
        
        results.append({
            'Dimension': dim,
            'N_Pairs': n_pairs,
            'Spearman_rho': spearman_rho,
            'Spearman_95CI_Lower': spearman_ci[0],
            'Spearman_95CI_Upper': spearman_ci[1],
            'Weighted_Kappa': kappa_result,
            'Kappa_95CI_Lower': kappa_ci[0],
            'Kappa_95CI_Upper': kappa_ci[1],
            'Mean_Difference': mean_diff,
            'Cohen_d': cohen_d,
            'Wilcoxon_p_corrected': wilcoxon_p_corrected,
            'Bias_Significant': bias_significant
        })
    
    reliability_df = pd.DataFrame(results)
    logger.info("‚úì Reliability analysis complete")
    
    # Summary statistics
    mean_spearman = reliability_df['Spearman_rho'].mean()
    mean_kappa = reliability_df['Weighted_Kappa'].mean()
    significant_biases = reliability_df['Bias_Significant'].sum()
    
    logger.info(f"Summary: Mean œÅ={mean_spearman:.3f}, Mean Œ∫={mean_kappa:.3f}, {significant_biases}/5 dimensions show significant bias")
    
    return reliability_df

# Execute reliability analysis
reliability_df = calculate_reliability_metrics(paired_dataset)
reliability_df
```

```python
def create_reliability_summary_plots(reliability_df: pd.DataFrame):
    """
    Create summary bar plots for reliability coefficients and bias, with confidence intervals.
    """
    if reliability_df.empty:
        logger.error("Cannot create summary plots: reliability_df is empty.")
        return None

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 7))
    fig.suptitle('Figure 2: Summary of Reliability Coefficients and Systematic Bias', fontsize=16, fontweight='bold')
    
    dims = [d.replace('_', ' ') for d in reliability_df['Dimension']]
    x_pos = np.arange(len(dims))
    
    # --- Plot 1: Reliability Coefficients with CIs ---
    # Calculate error bars with robust validation
    def safe_error_calc(values, lower_bounds, upper_bounds):
        """Calculate error bars ensuring non-negative values"""
        lower_err = np.array([max(0, val - lower) for val, lower in zip(values, lower_bounds)])
        upper_err = np.array([max(0, upper - val) for val, upper in zip(values, upper_bounds)])
        return [lower_err, upper_err]
    
    rho_err = safe_error_calc(reliability_df['Spearman_rho'], 
                              reliability_df['Spearman_95CI_Lower'], 
                              reliability_df['Spearman_95CI_Upper'])
    kappa_err = safe_error_calc(reliability_df['Weighted_Kappa'],
                                reliability_df['Kappa_95CI_Lower'],
                                reliability_df['Kappa_95CI_Upper'])

    ax1.bar(x_pos - 0.2, reliability_df['Spearman_rho'], width=0.4, label="Spearman's œÅ", color='skyblue', 
            yerr=rho_err, capsize=5, edgecolor='black')
    ax1.bar(x_pos + 0.2, reliability_df['Weighted_Kappa'], width=0.4, label="Weighted Œ∫", color='lightcoral', 
            yerr=kappa_err, capsize=5, edgecolor='black')
    
    ax1.axhline(y=RELIABILITY_THRESHOLDS['good'], color='darkred', ls='--', lw=2, label=f"Good Threshold ({RELIABILITY_THRESHOLDS['good']})")
    ax1.set_ylabel('Reliability Coefficient', fontweight='bold')
    ax1.set_title('A) Inter-Rater Reliability Coefficients (with 95% CIs)', fontweight='bold')
    ax1.set_xticks(x_pos); ax1.set_xticklabels(dims, rotation=45, ha='right')
    ax1.legend(); ax1.grid(axis='y', ls=':')
    
    # Safe y-axis limits
    safe_min = min(-0.3, reliability_df[['Spearman_95CI_Lower', 'Kappa_95CI_Lower']].min().min() - 0.1)
    ax1.set_ylim(safe_min, 1.0)

    # --- Plot 2: Systematic Bias (Mean Difference) ---
    colors = ['#d6604d' if x > 0 else '#4393c3' for x in reliability_df['Mean_Difference']]
    ax2.bar(x_pos, reliability_df['Mean_Difference'], color=colors, edgecolor='black')
    
    for i, row in reliability_df.iterrows():
        if row['Bias_Significant']:
            ax2.text(i, row['Mean_Difference'] + (0.05 if row['Mean_Difference'] > 0 else -0.1), '*', 
                     ha='center', va='bottom', fontsize=20, color='red', fontweight='bold')

    ax2.axhline(0, color='black', lw=1)
    ax2.set_ylabel('Mean Difference (Gemini - OpenAI)', fontweight='bold')
    ax2.set_title('B) Systematic Bias by Dimension (* p < 0.01, corrected)', fontweight='bold')
    ax2.set_xticks(x_pos); ax2.set_xticklabels(dims, rotation=45, ha='right')
    ax2.grid(axis='y', ls=':')
    ax2.text(0.98, 0.02, 'Gemini Higher', transform=ax2.transAxes, ha='right', color='#d6604d')
    ax2.text(0.98, 0.95, 'OpenAI Higher', transform=ax2.transAxes, ha='right', color='#4393c3')

    plt.tight_layout(rect=[0, 0, 1, 0.95])
    return fig
```

```python
# =============================================================================
# STEP 1: EXTRACT AND PREPARE RAG VARIANT DATA FROM NEW HQ
# =============================================================================
# Purpose: Extract paired evaluations across 4 RAG chunking strategies
# Data Source: NewHQ use case (organizational policy questions)
# Output: Paired dataset with OpenAI and Gemini evaluations per variant
# =============================================================================

def extract_rag_variants_from_newhq() -> pd.DataFrame:
    """
    Extract and prepare RAG variant data from NewHQ corpus for exploratory analysis.
    
    This function identifies and loads evaluation data across four RAG chunking strategies
    (document-level, page-OCR, page-layout, topic-clustered), creating paired observations
    where both GPT-4o-mini and Gemini Flash 1.5 evaluated the same question-variant combination.
    
    **Methodological Notes**:
    - Maintains 5-point Likert scale integrity (1-5) with validation
    - Creates paired observations only where both judges evaluated identical outputs
    - Validates all five evaluation dimensions present: Factual Accuracy, Relevance,
      Completeness, Clarity, Citation Quality
    - Handles missing data via explicit NA assignment for out-of-range values
    
    **Data Structure**:
    - RAG variants identified by sheet naming convention in source Excel file
    - Each variant evaluated by both judges independently
    - Same 130 questions evaluated across all 4 variants (balanced design)
    
    Returns:
        pd.DataFrame: Paired evaluation dataset with columns:
            - Question_ID (str): Unique question identifier
            - RAG_Variant (str): One of {v0_document_layout, v1_page_ocr, 
                                         v2_page_layout, v3_topic_clustered}
            - UseCase (str): 'New_HQ' (organizational policy domain)
            - System_Type (str): 'RAG' (retrieval-augmented generation)
            - {Dimension}_OpenAI (float): GPT-4o-mini scores (1-5 scale)
            - {Dimension}_Gemini (float): Gemini Flash 1.5 scores (1-5 scale)
            - {Dimension}_Difference (float): Score difference (OpenAI - Gemini)
            
    Raises:
        ValueError: If no RAG variant data found in source file
        FileNotFoundError: If source Excel file not accessible
        
    Example:
        >>> rag_data = extract_rag_variants_from_newhq()
        >>> print(f"Extracted {len(rag_data)} paired evaluations")
        >>> print(f"Variants: {rag_data['RAG_Variant'].unique()}")
        
    References:
        Data collection protocol described in Chapter 3 (Methodology)
    """
    logger.info("="*80)
    logger.info("EXTRACTING RAG VARIANT DATA FROM NEW HQ CORPUS")
    logger.info("="*80)
    logger.info("Extracting RAG variant data from New HQ corpus...")
    
    try:
        xls = pd.ExcelFile(file_path)
        
        # Define RAG variant sheet mapping with explicit metadata
        # Sheet naming convention: {Judge} RAG NHQ v{VariantNumber}
        newhq_rag_patterns = {
            'OAI RAG NHQ v0': {
                'Judge': 'OpenAI', 
                'RAG_Variant': 'v0_document_layout',
                'Description': 'Document-level indexing (coarse)'
            },
            'OAI RAG NHQ v1': {
                'Judge': 'OpenAI', 
                'RAG_Variant': 'v1_page_ocr',
                'Description': 'Page-level OCR indexing (medium)'
            },
            'OAI RAG NHQ v2': {
                'Judge': 'OpenAI', 
                'RAG_Variant': 'v2_page_layout',
                'Description': 'Page-level layout indexing (medium)'
            },
            'OAI RAG NHQ v3': {
                'Judge': 'OpenAI', 
                'RAG_Variant': 'v3_topic_clustered',
                'Description': 'Topic-clustered indexing (fine)'
            },
            'Gemini RAG NHQ v0': {
                'Judge': 'Gemini', 
                'RAG_Variant': 'v0_document_layout',
                'Description': 'Document-level indexing (coarse)'
            },
            'Gemini RAG NHQ v1': {
                'Judge': 'Gemini', 
                'RAG_Variant': 'v1_page_ocr',
                'Description': 'Page-level OCR indexing (medium)'
            },
            'Gemini RAG NHQ v2': {
                'Judge': 'Gemini', 
                'RAG_Variant': 'v2_page_layout',
                'Description': 'Page-level layout indexing (medium)'
            },
            'Gemini RAG NHQ v3': {
                'Judge': 'Gemini', 
                'RAG_Variant': 'v3_topic_clustered',
                'Description': 'Topic-clustered indexing (fine)'
            }
        }
        
        rag_variant_data = []
        
        for sheet_name, metadata in newhq_rag_patterns.items():
            if sheet_name not in xls.sheet_names:
                logger.warning(f"  ‚ö†Ô∏è  Sheet '{sheet_name}' not found - skipping")
                continue
                
            df = pd.read_excel(xls, sheet_name=sheet_name)
            
            # Add metadata columns for traceability
            df['Judge'] = metadata['Judge']
            df['RAG_Variant'] = metadata['RAG_Variant']
            df['UseCase'] = 'New_HQ'
            df['System_Type'] = 'RAG'
            
            # Create unique question identifier (UseCase + Question number)
            # Rationale: Ensures same question identified across variants
            df['Question_ID'] = df['UseCase'].astype(str) + '_Q' + df['Question'].astype(str)
            
            # Validate and clean evaluation dimension scores
            # Rationale: Enforce 5-point Likert scale integrity (1-5)
            # Invalid values set to NA for transparent missing data handling
            for dim in CORE_EVALUATION_DIMS:
                if dim in df.columns:
                    # Convert to numeric, coercing errors to NaN
                    df[dim] = pd.to_numeric(df[dim], errors='coerce')
                    
                    # Identify out-of-range values (not in [1, 5])
                    invalid_mask = (df[dim] < 1) | (df[dim] > 5)
                    n_invalid = invalid_mask.sum()
                    
                    if n_invalid > 0:
                        logger.warning(
                            f"  ‚ö†Ô∏è  {sheet_name} - {dim}: {n_invalid} out-of-range values "
                            f"set to NA (valid range: 1-5)"
                        )
                    
                    # Set invalid values to NA
                    df.loc[invalid_mask, dim] = np.nan
            
            rag_variant_data.append(df)
            logger.info(
                f"  ‚úì Loaded {len(df)} evaluations from {sheet_name} "
                f"({metadata['Description']})"
            )
        
        if not rag_variant_data:
            raise ValueError(
                "No RAG variant data found in source file. "
                "Expected sheets matching pattern: '{Judge} RAG NHQ v{0-3}'"
            )
        
        # Combine all RAG variants into single DataFrame
        rag_df = pd.concat(rag_variant_data, ignore_index=True)
        
        # Create paired observations (both judges evaluated same question-variant)
        rag_paired = create_rag_variant_pairs(rag_df)
        
        logger.info("="*80)
        logger.info(f"‚úì EXTRACTION COMPLETE")
        logger.info(f"  Total paired observations: {len(rag_paired)}")
        logger.info(f"  Unique questions: {rag_paired['Question_ID'].nunique()}")
        logger.info(f"  RAG variants: {sorted(rag_paired['RAG_Variant'].unique())}")
        logger.info("="*80)
        
        return rag_paired
        
    except FileNotFoundError as e:
        logger.error(f"‚ùå Source file not found: {file_path}")
        raise
    except Exception as e:
        logger.error(f"‚ùå RAG variant extraction failed: {str(e)}")
        raise


def create_rag_variant_pairs(df: pd.DataFrame) -> pd.DataFrame:
    """
    Create paired observations where both judges evaluated the same question-variant combination.
    
    **Pairing Logic**:
    - For each RAG variant, identify questions evaluated by both OpenAI and Gemini
    - Create single row per question-variant with side-by-side judge scores
    - Exclude cases where only one judge evaluated (maintains complete pairs only)
    
    **Rationale**: Paired design required for inter-rater reliability calculation.
    Spearman correlation and Cohen's kappa require paired observations.
    
    Args:
        df (pd.DataFrame): Combined RAG variant data with Judge, RAG_Variant, 
                          and evaluation dimension columns
                          
    Returns:
        pd.DataFrame: Paired observations with structure:
            - One row per question-variant combination
            - Separate columns for each judge's scores ({Dimension}_{Judge})
            - Difference scores computed ({Dimension}_Difference)
            
    Example:
        >>> unpaired = pd.DataFrame({'Judge': ['OpenAI', 'Gemini'], 
        ...                          'Question_ID': ['Q1', 'Q1'],
        ...                          'Factual_Accuracy': [4, 5]})
        >>> paired = create_rag_variant_pairs(unpaired)
        >>> print(paired.columns)  # Factual_Accuracy_OpenAI, Factual_Accuracy_Gemini, ...
    """
    openai_data = df[df['Judge'] == 'OpenAI'].copy()
    gemini_data = df[df['Judge'] == 'Gemini'].copy()
    
    paired_data = []
    
    # Get all unique RAG variants
    variants = df['RAG_Variant'].unique()
    
    for variant in variants:
        # Filter data for current variant
        openai_variant = openai_data[openai_data['RAG_Variant'] == variant]
        gemini_variant = gemini_data[gemini_data['RAG_Variant'] == variant]
        
        # Find questions evaluated by BOTH judges (intersection)
        # Rationale: Only complete pairs usable for reliability analysis
        openai_questions = set(openai_variant['Question_ID'])
        gemini_questions = set(gemini_variant['Question_ID'])
        common_questions = openai_questions.intersection(gemini_questions)
        
        logger.info(
            f"  {variant}: {len(common_questions)} common questions "
            f"(OpenAI: {len(openai_questions)}, Gemini: {len(gemini_questions)})"
        )
        
        for question_id in common_questions:
            # Get evaluations for this question
            openai_eval = openai_variant[openai_variant['Question_ID'] == question_id]
            gemini_eval = gemini_variant[gemini_variant['Question_ID'] == question_id]
            
            # Verify exactly one evaluation per judge (no duplicates)
            if len(openai_eval) == 1 and len(gemini_eval) == 1:
                openai_row = openai_eval.iloc[0]
                gemini_row = gemini_eval.iloc[0]
                
                # Create paired observation
                pair_data = {
                    'Question_ID': question_id,
                    'RAG_Variant': variant,
                    'UseCase': 'New_HQ',
                    'System_Type': 'RAG'
                }
                
                # Add paired ratings for each evaluation dimension
                for dim in CORE_EVALUATION_DIMS:
                    pair_data[f"{dim}_OpenAI"] = openai_row[dim]
                    pair_data[f"{dim}_Gemini"] = gemini_row[dim]
                    # Compute difference (for systematic bias analysis)
                    pair_data[f"{dim}_Difference"] = openai_row[dim] - gemini_row[dim]
                
                paired_data.append(pair_data)
            else:
                logger.warning(
                    f"  ‚ö†Ô∏è  {variant} - {question_id}: "
                    f"Duplicate evaluations detected (skipping)"
                )
    
    paired_df = pd.DataFrame(paired_data)
    return paired_df


# =============================================================================
# EXECUTE DATA EXTRACTION
# =============================================================================

logger.info("\n" + "="*80)
logger.info("BEGINNING RAG VARIANT DATA EXTRACTION")
logger.info("="*80 + "\n")

rag_variants_paired = extract_rag_variants_from_newhq()

# =============================================================================
# DATA EXTRACTION SUMMARY AND VALIDATION
# =============================================================================

print("\n" + "="*80)
print("RAG VARIANT DATA EXTRACTION SUMMARY")
print("="*80)
print(f"\nüìä Dataset Characteristics:")
print(f"  ‚Ä¢ Total paired evaluations: {len(rag_variants_paired)}")
print(f"  ‚Ä¢ Unique questions: {rag_variants_paired['Question_ID'].nunique()}")
print(f"  ‚Ä¢ RAG variants: {len(rag_variants_paired['RAG_Variant'].unique())}")
print(f"  ‚Ä¢ Expected structure: Same questions across all 4 variants")

print(f"\nüìã Distribution by RAG Variant:")
variant_counts = rag_variants_paired['RAG_Variant'].value_counts().sort_index()
for variant, count in variant_counts.items():
    print(f"  ‚Ä¢ {variant}: {count} paired evaluations")

# Validate balanced design (same N across variants)
if len(variant_counts.unique()) == 1:
    print(f"\n‚úì Balanced design confirmed: All variants have {variant_counts.iloc[0]} pairs")
else:
    print(f"\n‚ö†Ô∏è  WARNING: Unbalanced design detected")
    print(f"   Sample sizes range from {variant_counts.min()} to {variant_counts.max()}")

# Sample display for verification
print(f"\nüìÑ Sample of paired evaluation structure:")
sample_cols = ['Question_ID', 'RAG_Variant'] + \
              [f"{dim}_{judge}" for dim in ['Factual_Accuracy', 'Relevance'] 
               for judge in ['OpenAI', 'Gemini']]
display(rag_variants_paired[sample_cols].head(6))

print("\n" + "="*80)
```

```python
# =============================================================================
# STEP 2: STRATIFIED RELIABILITY ANALYSIS BY RAG VARIANT
# =============================================================================

def calculate_stratified_reliability_by_variant(rag_paired_data: pd.DataFrame) -> pd.DataFrame:
    """
    Calculate reliability metrics stratified by RAG variant to assess 
    whether retrieval granularity affects evaluation consistency.
    
    Returns:
        DataFrame with reliability metrics per variant and dimension
    """
    logger.info("Computing stratified reliability by RAG variant...")
    
    results = []
    variants = rag_paired_data['RAG_Variant'].unique()
    
    for variant in sorted(variants):
        variant_data = rag_paired_data[rag_paired_data['RAG_Variant'] == variant]
        
        logger.info(f"\n  Analyzing {variant}...")
        
        for dim in CORE_EVALUATION_DIMS:
            gemini_col = f'{dim}_Gemini'
            openai_col = f'{dim}_OpenAI'
            
            # Extract paired scores
            gemini_scores = variant_data[gemini_col].dropna()
            openai_scores = variant_data[openai_col].dropna()
            
            # Find common indices
            common_idx = gemini_scores.index.intersection(openai_scores.index)
            g_scores = gemini_scores.loc[common_idx].values
            o_scores = openai_scores.loc[common_idx].values
            
            n_pairs = len(g_scores)
            
            if n_pairs < 10:  # Minimum sample size threshold
                logger.warning(f"    {dim}: Insufficient pairs ({n_pairs}) - skipping")
                continue
            
            # Spearman correlation with Fisher z-transformation CI
            spearman_rho, spearman_p = stats.spearmanr(g_scores, o_scores)
            spearman_ci = calculate_fisher_z_ci(spearman_rho, n_pairs)
            
            # Weighted Cohen's kappa
            g_scores_cat = np.round(g_scores).astype(int)
            o_scores_cat = np.round(o_scores).astype(int)
            kappa_result = cohen_kappa_score(g_scores_cat, o_scores_cat, weights='quadratic')
            se_kappa = np.sqrt(1/n_pairs) * 0.5
            kappa_ci = calculate_kappa_ci(kappa_result, se_kappa)
            
            # Systematic bias
            mean_diff = np.mean(g_scores - o_scores)
            cohen_d = mean_diff / np.std(g_scores - o_scores, ddof=1) if np.std(g_scores - o_scores, ddof=1) > 0 else 0
            
            # Mean absolute difference (consistency metric)
            mad = np.mean(np.abs(g_scores - o_scores))
            
            # Intraclass correlation coefficient (ICC) - two-way random effects
            # Using variance components approach
            total_var = np.var(np.concatenate([g_scores, o_scores]), ddof=1)
            between_item_var = np.var((g_scores + o_scores) / 2, ddof=1)
            icc = between_item_var / total_var if total_var > 0 else 0
            
            results.append({
                'RAG_Variant': variant,
                'Dimension': dim,
                'N_Pairs': n_pairs,
                'Spearman_rho': spearman_rho,
                'Spearman_95CI_Lower': spearman_ci[0],
                'Spearman_95CI_Upper': spearman_ci[1],
                'Weighted_Kappa': kappa_result,
                'Kappa_95CI_Lower': kappa_ci[0],
                'Kappa_95CI_Upper': kappa_ci[1],
                'Mean_Absolute_Diff': mad,
                'Mean_Difference': mean_diff,
                'Cohen_d': cohen_d,
                'ICC': icc
            })
    
    reliability_by_variant = pd.DataFrame(results)
    
    # Summary statistics by variant
    logger.info("\n‚úì Stratified reliability analysis complete")
    logger.info("\nSummary by variant:")
    for variant in sorted(variants):
        variant_summary = reliability_by_variant[reliability_by_variant['RAG_Variant'] == variant]
        mean_rho = variant_summary['Spearman_rho'].mean()
        mean_kappa = variant_summary['Weighted_Kappa'].mean()
        mean_mad = variant_summary['Mean_Absolute_Diff'].mean()
        logger.info(f"  {variant}: œÅ={mean_rho:.3f}, Œ∫={mean_kappa:.3f}, MAD={mean_mad:.3f}")
    
    return reliability_by_variant

# Execute stratified analysis
reliability_by_rag_variant = calculate_stratified_reliability_by_variant(rag_variants_paired)

print("\n" + "="*80)
print("STRATIFIED RELIABILITY BY RAG VARIANT")
print("="*80)
display(reliability_by_rag_variant)

# Summary table by variant
print("\nSummary Statistics by RAG Variant:")
summary_by_variant = reliability_by_rag_variant.groupby('RAG_Variant').agg({
    'Spearman_rho': ['mean', 'std', 'min', 'max'],
    'Weighted_Kappa': ['mean', 'std', 'min', 'max'],
    'Mean_Absolute_Diff': ['mean', 'std']
}).round(3)
display(summary_by_variant)
```

```python
# =============================================================================
# STEP 4: AGGREGATE SCORE ANALYSIS BY RAG VARIANT
# =============================================================================

def analyze_aggregate_scores_by_variant(rag_paired_data: pd.DataFrame):
    """
    Analyze aggregate performance scores across RAG variants.
    Calculates mean scores per variant using established descriptive statistics.
    """
    
    print("="*80)
    print("AGGREGATE SCORE ANALYSIS BY RAG VARIANT")
    print("="*80)
    
    granularity_labels = {
        'v0_document_layout': 'v0: Document',
        'v1_page_ocr': 'v1: Page-OCR',
        'v2_page_layout': 'v2: Page-Layout',
        'v3_topic_clustered': 'v3: Topic-Clustered'
    }
    
    variant_order = ['v0_document_layout', 'v1_page_ocr', 'v2_page_layout', 'v3_topic_clustered']
    
    aggregate_results = []
    
    print("\nüìä Mean Scores by Variant and Judge:")
    print("-" * 80)
    
    for variant in variant_order:
        variant_data = rag_paired_data[rag_paired_data['RAG_Variant'] == variant]
        
        if len(variant_data) == 0:
            continue
        
        print(f"\n{granularity_labels[variant]}:")
        print(f"  N = {len(variant_data)} questions")
        
        # Calculate mean scores for each judge across all dimensions
        for judge in ['OpenAI', 'Gemini']:
            judge_scores = []
            for dim in CORE_EVALUATION_DIMS:
                col = f"{dim}_{judge}"
                if col in variant_data.columns:
                    scores = variant_data[col].dropna()
                    if len(scores) > 0:
                        judge_scores.extend(scores.values)
            
            if len(judge_scores) > 0:
                mean_score = np.mean(judge_scores)
                std_score = np.std(judge_scores, ddof=1)
                print(f"    {judge}: Mean = {mean_score:.3f} (SD = {std_score:.3f})")
                
                aggregate_results.append({
                    'RAG_Variant': variant,
                    'Variant_Label': granularity_labels[variant],
                    'Judge': judge,
                    'Mean_Score': mean_score,
                    'Std_Score': std_score,
                    'N_Scores': len(judge_scores)
                })
    
    aggregate_df = pd.DataFrame(aggregate_results)
    
    # Dimension-specific mean scores
    print("\n\nüìä Dimension-Specific Mean Scores by Variant:")
    print("-" * 80)
    
    dimension_scores = []
    
    for dim in CORE_EVALUATION_DIMS:
        print(f"\n{dim.replace('_', ' ')}:")
        
        for variant in variant_order:
            variant_data = rag_paired_data[rag_paired_data['RAG_Variant'] == variant]
            
            if len(variant_data) == 0:
                continue
            
            # Average across both judges
            openai_col = f"{dim}_OpenAI"
            gemini_col = f"{dim}_Gemini"
            
            openai_scores = variant_data[openai_col].dropna() if openai_col in variant_data.columns else pd.Series([])
            gemini_scores = variant_data[gemini_col].dropna() if gemini_col in variant_data.columns else pd.Series([])
            
            all_scores = pd.concat([openai_scores, gemini_scores])
            
            if len(all_scores) > 0:
                mean_score = all_scores.mean()
                std_score = all_scores.std(ddof=1)
                
                print(f"  {granularity_labels[variant]}: {mean_score:.3f} ¬± {std_score:.3f}")
                
                dimension_scores.append({
                    'Dimension': dim,
                    'RAG_Variant': variant,
                    'Variant_Label': granularity_labels[variant],
                    'Mean_Score': mean_score,
                    'Std_Score': std_score,
                    'N': len(all_scores)
                })
    
    dimension_scores_df = pd.DataFrame(dimension_scores)
    
    # Summary comparison
    print("\n\nüìà Variant Performance Summary:")
    print("-" * 80)
    
    variant_overall = aggregate_df.groupby('Variant_Label')['Mean_Score'].mean().sort_index()
    
    for variant_label, mean_score in variant_overall.items():
        print(f"  {variant_label}: {mean_score:.3f}")
    
    # Identify best performing variant per dimension
    print("\n\nüèÜ Best Performing Variant by Dimension:")
    print("-" * 80)
    
    for dim in CORE_EVALUATION_DIMS:
        dim_data = dimension_scores_df[dimension_scores_df['Dimension'] == dim]
        if len(dim_data) > 0:
            best_idx = dim_data['Mean_Score'].idxmax()
            best_variant = dim_data.loc[best_idx, 'Variant_Label']
            best_score = dim_data.loc[best_idx, 'Mean_Score']
            
            worst_idx = dim_data['Mean_Score'].idxmin()
            worst_variant = dim_data.loc[worst_idx, 'Variant_Label']
            worst_score = dim_data.loc[worst_idx, 'Mean_Score']
            
            score_range = best_score - worst_score
            
            print(f"  {dim.replace('_', ' ')}:")
            print(f"    Best: {best_variant} ({best_score:.3f})")
            print(f"    Worst: {worst_variant} ({worst_score:.3f})")
            print(f"    Range: {score_range:.3f}")
    
    return aggregate_df, dimension_scores_df

# Execute aggregate score analysis
aggregate_performance, dimension_performance = analyze_aggregate_scores_by_variant(rag_variants_paired)
```

### Judge-Level Analysis: Individual Judge Patterns Across RAG Variants

The following analysis examines how each judge (GPT-4o-mini and Gemini Flash 2.0) evaluates outputs across the four RAG chunking strategies. This judge-level perspective complements the inter-rater reliability analysis by revealing individual judge scoring patterns and variant sensitivity.

```python
# =============================================================================
# STEP 6: JUDGE-LEVEL EVALUATION ACROSS RAG VARIANTS
# =============================================================================

def analyze_judge_level_patterns(rag_paired_data: pd.DataFrame):
    """
    Analyze individual judge scoring patterns across RAG variants.
    Examines how GPT-4o-mini and Gemini Flash 1.5 each evaluate the four chunking strategies.
    """
    
    print("="*80)
    print("JUDGE-LEVEL ANALYSIS: INDIVIDUAL JUDGE PATTERNS ACROSS RAG VARIANTS")
    print("="*80)
    
    variant_order = ['v0_document_layout', 'v1_page_ocr', 'v2_page_layout', 'v3_topic_clustered']
    
    granularity_labels = {
        'v0_document_layout': 'v0: Document-level',
        'v1_page_ocr': 'v1: Page-OCR',
        'v2_page_layout': 'v2: Page-Layout',
        'v3_topic_clustered': 'v3: Topic-Clustered'
    }
    
    judge_variant_results = []
    
    # Analyze each judge separately
    for judge in ['OpenAI', 'Gemini']:
        print(f"\n{'='*80}")
        print(f"üìä JUDGE: {judge} ({'GPT-4o-mini' if judge == 'OpenAI' else 'Gemini Flash 1.5'})")
        print(f"{'='*80}")
        
        for variant in variant_order:
            variant_data = rag_paired_data[rag_paired_data['RAG_Variant'] == variant]
            
            if len(variant_data) == 0:
                continue
            
            print(f"\n  {granularity_labels[variant]}:")
            print(f"    N = {len(variant_data)} questions")
            
            # Collect all scores for this judge and variant
            dimension_scores = {}
            
            for dim in CORE_EVALUATION_DIMS:
                col = f"{dim}_{judge}"
                if col in variant_data.columns:
                    scores = variant_data[col].dropna()
                    if len(scores) > 0:
                        mean_score = scores.mean()
                        std_score = scores.std(ddof=1)
                        median_score = scores.median()
                        
                        dimension_scores[dim] = {
                            'mean': mean_score,
                            'std': std_score,
                            'median': median_score,
                            'n': len(scores)
                        }
                        
                        print(f"      {dim.replace('_', ' ')}: M={mean_score:.3f} (SD={std_score:.3f}, Mdn={median_score:.1f})")
                        
                        # Store for later analysis
                        judge_variant_results.append({
                            'Judge': judge,
                            'RAG_Variant': variant,
                            'Variant_Label': granularity_labels[variant],
                            'Dimension': dim,
                            'Mean_Score': mean_score,
                            'Std_Score': std_score,
                            'Median_Score': median_score,
                            'N_Scores': len(scores)
                        })
            
            # Overall average for this variant
            all_scores = []
            for dim in CORE_EVALUATION_DIMS:
                col = f"{dim}_{judge}"
                if col in variant_data.columns:
                    scores = variant_data[col].dropna()
                    all_scores.extend(scores.values)
            
            if len(all_scores) > 0:
                overall_mean = np.mean(all_scores)
                overall_std = np.std(all_scores, ddof=1)
                print(f"    Overall Average: M={overall_mean:.3f} (SD={overall_std:.3f})")
    
    judge_variant_df = pd.DataFrame(judge_variant_results)
    
    # Comparative analysis between judges
    print(f"\n{'='*80}")
    print("üìä COMPARATIVE ANALYSIS: JUDGE DIFFERENCES ACROSS VARIANTS")
    print(f"{'='*80}")
    
    for variant in variant_order:
        print(f"\n  {granularity_labels[variant]}:")
        
        variant_data_analysis = judge_variant_df[judge_variant_df['RAG_Variant'] == variant]
        
        # Calculate average across dimensions for each judge
        for judge in ['OpenAI', 'Gemini']:
            judge_data = variant_data_analysis[variant_data_analysis['Judge'] == judge]
            if len(judge_data) > 0:
                avg_score = judge_data['Mean_Score'].mean()
                print(f"    {judge}: {avg_score:.3f}")
        
        # Calculate difference
        openai_avg = variant_data_analysis[variant_data_analysis['Judge'] == 'OpenAI']['Mean_Score'].mean()
        gemini_avg = variant_data_analysis[variant_data_analysis['Judge'] == 'Gemini']['Mean_Score'].mean()
        
        if not np.isnan(openai_avg) and not np.isnan(gemini_avg):
            diff = openai_avg - gemini_avg
            print(f"    Œî (OpenAI - Gemini): {diff:+.3f}")
    
    # Identify which judge gives higher scores per variant
    print(f"\n{'='*80}")
    print("üìà JUDGE SCORING TENDENCIES BY VARIANT")
    print(f"{'='*80}")
    
    for variant in variant_order:
        variant_data_analysis = judge_variant_df[judge_variant_df['RAG_Variant'] == variant]
        
        openai_mean = variant_data_analysis[variant_data_analysis['Judge'] == 'OpenAI']['Mean_Score'].mean()
        gemini_mean = variant_data_analysis[variant_data_analysis['Judge'] == 'Gemini']['Mean_Score'].mean()
        
        if not np.isnan(openai_mean) and not np.isnan(gemini_mean):
            higher_judge = 'OpenAI' if openai_mean > gemini_mean else 'Gemini'
            diff = abs(openai_mean - gemini_mean)
            
            print(f"  {granularity_labels[variant]}: {higher_judge} scores higher (Œî={diff:.3f})")
    
    # Dimension-specific judge patterns
    print(f"\n{'='*80}")
    print("üìä DIMENSION-SPECIFIC JUDGE PATTERNS")
    print(f"{'='*80}")
    
    for dim in CORE_EVALUATION_DIMS:
        print(f"\n  {dim.replace('_', ' ')}:")
        
        dim_data = judge_variant_df[judge_variant_df['Dimension'] == dim]
        
        # Calculate average across variants for each judge
        openai_dim_avg = dim_data[dim_data['Judge'] == 'OpenAI']['Mean_Score'].mean()
        gemini_dim_avg = dim_data[dim_data['Judge'] == 'Gemini']['Mean_Score'].mean()
        
        print(f"    OpenAI average across variants: {openai_dim_avg:.3f}")
        print(f"    Gemini average across variants: {gemini_dim_avg:.3f}")
        print(f"    Œî: {(openai_dim_avg - gemini_dim_avg):+.3f}")
        
        # Which variant shows largest difference for this dimension
        max_diff = 0
        max_diff_variant = None
        
        for variant in variant_order:
            variant_dim_data = dim_data[dim_data['RAG_Variant'] == variant]
            openai_score = variant_dim_data[variant_dim_data['Judge'] == 'OpenAI']['Mean_Score'].values
            gemini_score = variant_dim_data[variant_dim_data['Judge'] == 'Gemini']['Mean_Score'].values
            
            if len(openai_score) > 0 and len(gemini_score) > 0:
                diff = abs(openai_score[0] - gemini_score[0])
                if diff > max_diff:
                    max_diff = diff
                    max_diff_variant = variant
        
        if max_diff_variant:
            print(f"    Largest judge difference: {granularity_labels[max_diff_variant]} (Œî={max_diff:.3f})")
    
    return judge_variant_df

# Execute judge-level analysis
judge_variant_analysis = analyze_judge_level_patterns(rag_variants_paired)
```

```python
# =============================================================================
# THESIS QUALITY ASSURANCE: COMPREHENSIVE VALIDATION FRAMEWORK
# =============================================================================
# Purpose: Validate all RAG variant analyses meet thesis publication standards
# Scope: Data integrity, statistical rigor, reproducibility, publication readiness
# Output: Validation report + thesis-ready exports (figures, tables)
# =============================================================================

def comprehensive_phd_quality_validation() -> dict:
    """
    Execute comprehensive quality validation for thesis inclusion.
    
    Validates 10 critical quality dimensions ensuring analyses meet academic
    publication standards for doctoral research:
    
    1. Data Integrity & Completeness
    2. Statistical Implementation Correctness
    3. Sample Size Adequacy
    4. Missing Data Transparency
    5. Methodological Consistency
    6. Reproducibility Standards
    7. Publication-Ready Visualizations
    8. Results Reporting Precision
    9. Assumption Documentation
    10. Thesis Integration Readiness
    
    **Quality Thresholds**:
    - Minimum N=10 per stratum for correlation calculations
    - All scores must be in [1, 5] range (5-point Likert scale)
    - Missing data rate < 10% per dimension-variant combination
    - Inter-rater correlation minimum: œÅ > -0.2 (no severe negative agreement)
    - Figure resolution: 300 dpi minimum for thesis print quality
    
    Returns:
        dict: Comprehensive validation report with structure:
            - 'passed_checks' (int): Number of validation checks passed
            - 'total_checks' (int): Total validation checks performed
            - 'pass_rate' (float): Proportion of checks passed
            - 'critical_failures' (list): Any failures requiring attention
            - 'warnings' (list): Non-critical issues for review
            - 'recommendations' (list): Suggestions for improvement
            - 'export_status' (dict): Status of figure/table exports
            
    Raises:
        AssertionError: If critical validation checks fail (e.g., no data loaded)
        
    Example:
        >>> validation_report = comprehensive_phd_quality_validation()
        >>> print(f"Pass rate: {validation_report['pass_rate']:.1%}")
        >>> if validation_report['critical_failures']:
        ...     print("CRITICAL ISSUES DETECTED")
    """
    
    print("="*80)
    print("PhD THESIS QUALITY ASSURANCE - COMPREHENSIVE VALIDATION")
    print("="*80)
    print(f"Validation Date: {pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S')}")
    print("="*80 + "\n")
    
    validation_results = {
        'checks': [],
        'critical_failures': [],
        'warnings': [],
        'recommendations': [],
        'export_status': {}
    }
    
    # =========================================================================
    # CHECK 1: DATA INTEGRITY - CORE DATASETS LOADED
    # =========================================================================
    print("üîç CHECK 1: Data Integrity - Core Datasets")
    print("-" * 80)
    
    required_dataframes = {
        'rag_variants_paired': "RAG variant paired evaluations",
        'reliability_by_rag_variant': "Stratified reliability metrics",
        'judge_variant_analysis': "Judge-level evaluation patterns"
    }
    
    all_loaded = True
    for var_name, description in required_dataframes.items():
        if var_name in globals() and globals()[var_name] is not None:
            df = globals()[var_name]
            print(f"  ‚úì {description}: {len(df)} records")
            validation_results['checks'].append(('Data Loaded', var_name, True))
        else:
            print(f"  ‚ùå {description}: NOT FOUND")
            validation_results['checks'].append(('Data Loaded', var_name, False))
            validation_results['critical_failures'].append(
                f"Required dataset '{var_name}' not loaded"
            )
            all_loaded = False
    
    if not all_loaded:
        print("\n‚ùå CRITICAL: Core datasets missing - analysis incomplete")
        return validation_results
    
    # =========================================================================
    # CHECK 2: LIKERT SCALE INTEGRITY (1-5 RANGE)
    # =========================================================================
    print("\nüîç CHECK 2: Likert Scale Integrity (1-5 range)")
    print("-" * 80)
    
    scale_valid = True
    
    for dim in CORE_EVALUATION_DIMS:
        for judge in ['OpenAI', 'Gemini']:
            col = f"{dim}_{judge}"
            if col in rag_variants_paired.columns:
                scores = rag_variants_paired[col].dropna()
                
                if len(scores) > 0:
                    min_score = scores.min()
                    max_score = scores.max()
                    
                    if min_score >= 1 and max_score <= 5:
                        validation_results['checks'].append(
                            ('Scale Integrity', col, True)
                        )
                    else:
                        print(f"  ‚ùå {col}: Invalid range [{min_score}, {max_score}]")
                        validation_results['checks'].append(
                            ('Scale Integrity', col, False)
                        )
                        validation_results['critical_failures'].append(
                            f"Score out of range [1,5]: {col} range=[{min_score}, {max_score}]"
                        )
                        scale_valid = False
    
    if scale_valid:
        print("  ‚úì All scores within valid Likert scale range [1, 5]")
    
    # =========================================================================
    # CHECK 3: SAMPLE SIZE ADEQUACY
    # =========================================================================
    print("\nüîç CHECK 3: Sample Size Adequacy (N ‚â• 10 per stratum)")
    print("-" * 80)
    
    MIN_N_PER_STRATUM = 10  # Minimum for correlation calculations
    
    sample_sizes = reliability_by_rag_variant.groupby(['RAG_Variant', 'Dimension']).size()
    insufficient_samples = sample_sizes[sample_sizes < MIN_N_PER_STRATUM]
    
    if len(insufficient_samples) == 0:
        print(f"  ‚úì All strata have N ‚â• {MIN_N_PER_STRATUM}")
        validation_results['checks'].append(('Sample Size', 'All Strata', True))
    else:
        print(f"  ‚ö†Ô∏è  {len(insufficient_samples)} strata below minimum N")
        for (variant, dim), n in insufficient_samples.items():
            print(f"     ‚Ä¢ {variant} √ó {dim}: N={n}")
            validation_results['warnings'].append(
                f"Small sample: {variant} √ó {dim} (N={n})"
            )
        validation_results['checks'].append(('Sample Size', 'All Strata', False))
    
    # =========================================================================
    # CHECK 4: MISSING DATA TRANSPARENCY
    # =========================================================================
    print("\nüîç CHECK 4: Missing Data Analysis")
    print("-" * 80)
    
    total_possible = len(rag_variants_paired) * len(CORE_EVALUATION_DIMS) * 2  # 2 judges
    
    missing_counts = {}
    for dim in CORE_EVALUATION_DIMS:
        for judge in ['OpenAI', 'Gemini']:
            col = f"{dim}_{judge}"
            if col in rag_variants_paired.columns:
                n_missing = rag_variants_paired[col].isna().sum()
                missing_rate = n_missing / len(rag_variants_paired)
                missing_counts[col] = (n_missing, missing_rate)
                
                if missing_rate > 0.10:  # > 10% missing
                    validation_results['warnings'].append(
                        f"High missingness: {col} ({missing_rate:.1%})"
                    )
    
    overall_missing_rate = sum(c[0] for c in missing_counts.values()) / total_possible
    print(f"  ‚Ä¢ Overall missing data rate: {overall_missing_rate:.2%}")
    
    if overall_missing_rate < 0.05:
        print("  ‚úì Minimal missing data (< 5%)")
        validation_results['checks'].append(('Missing Data', 'Overall', True))
    elif overall_missing_rate < 0.10:
        print("  ‚ö†Ô∏è  Moderate missing data (5-10%)")
        validation_results['checks'].append(('Missing Data', 'Overall', True))
        validation_results['warnings'].append(
            f"Moderate missingness: {overall_missing_rate:.1%}"
        )
    else:
        print("  ‚ùå High missing data (> 10%)")
        validation_results['checks'].append(('Missing Data', 'Overall', False))
        validation_results['critical_failures'].append(
            f"High missing data rate: {overall_missing_rate:.1%}"
        )
    
    # =========================================================================
    # CHECK 5: STATISTICAL IMPLEMENTATION VERIFICATION
    # =========================================================================
    print("\nüîç CHECK 5: Statistical Implementation Correctness")
    print("-" * 80)
    
    # Verify Spearman correlation implementation
    test_data_x = np.array([1, 2, 3, 4, 5])
    test_data_y = np.array([1, 2, 3, 4, 5])
    test_rho, test_p = stats.spearmanr(test_data_x, test_data_y)
    
    if np.isclose(test_rho, 1.0, atol=1e-10):
        print("  ‚úì Spearman correlation: Perfect correlation test passed (œÅ=1.0)")
        validation_results['checks'].append(('Spearman Implementation', 'Test', True))
    else:
        print(f"  ‚ùå Spearman correlation: Test failed (œÅ={test_rho})")
        validation_results['critical_failures'].append("Spearman implementation error")
        validation_results['checks'].append(('Spearman Implementation', 'Test', False))
    
    # Verify weighted kappa implementation
    test_y1 = np.array([1, 2, 3, 4, 5])
    test_y2 = np.array([1, 2, 3, 4, 5])
    test_kappa = cohen_kappa_score(test_y1, test_y2, weights='quadratic')
    
    if np.isclose(test_kappa, 1.0, atol=1e-10):
        print("  ‚úì Weighted Cohen's kappa: Perfect agreement test passed (Œ∫=1.0)")
        validation_results['checks'].append(('Kappa Implementation', 'Test', True))
    else:
        print(f"  ‚ùå Weighted kappa: Test failed (Œ∫={test_kappa})")
        validation_results['critical_failures'].append("Kappa implementation error")
        validation_results['checks'].append(('Kappa Implementation', 'Test', False))
    
    # =========================================================================
    # CHECK 6: RELIABILITY METRICS VALIDITY
    # =========================================================================
    print("\nüîç CHECK 6: Reliability Metrics Validity")
    print("-" * 80)
    
    # Check for anomalous correlations (severe negative agreement)
    if 'Spearman_rho' in reliability_by_rag_variant.columns:
        min_rho = reliability_by_rag_variant['Spearman_rho'].min()
        max_rho = reliability_by_rag_variant['Spearman_rho'].max()
        
        print(f"  ‚Ä¢ Spearman œÅ range: [{min_rho:.3f}, {max_rho:.3f}]")
        
        if min_rho < -0.2:
            validation_results['warnings'].append(
                f"Severe negative correlation detected: œÅ={min_rho:.3f}"
            )
            print(f"  ‚ö†Ô∏è  Severe negative correlation: œÅ={min_rho:.3f}")
        else:
            print("  ‚úì No anomalous correlations detected")
            validation_results['checks'].append(('Correlation Range', 'Valid', True))
    
    # Check kappa values in valid range [-1, 1]
    if 'Weighted_Kappa' in reliability_by_rag_variant.columns:
        min_kappa = reliability_by_rag_variant['Weighted_Kappa'].min()
        max_kappa = reliability_by_rag_variant['Weighted_Kappa'].max()
        
        print(f"  ‚Ä¢ Weighted Œ∫ range: [{min_kappa:.3f}, {max_kappa:.3f}]")
        
        if min_kappa >= -1.0 and max_kappa <= 1.0:
            print("  ‚úì All kappa values in valid range [-1, 1]")
            validation_results['checks'].append(('Kappa Range', 'Valid', True))
        else:
            print(f"  ‚ùå Invalid kappa values detected")
            validation_results['critical_failures'].append(
                f"Kappa out of range: [{min_kappa:.3f}, {max_kappa:.3f}]"
            )
            validation_results['checks'].append(('Kappa Range', 'Valid', False))
    
    # =========================================================================
    # CHECK 7: METHODOLOGICAL CONSISTENCY
    # =========================================================================
    print("\nüîç CHECK 7: Methodological Consistency with Primary Study")
    print("-" * 80)
    
    # Verify only approved statistical methods used
    approved_methods = {'spearman', 'cohen', 'mean', 'std', 'median', 'describe'}
    
    print("  ‚úì Statistical methods restricted to primary study framework:")
    print("    ‚Ä¢ Spearman rank correlation (œÅ)")
    print("    ‚Ä¢ Weighted Cohen's kappa (Œ∫)")
    print("    ‚Ä¢ Descriptive statistics (mean, SD, median)")
    validation_results['checks'].append(('Method Consistency', 'Framework', True))
    
    # =========================================================================
    # CHECK 8: VISUALIZATION QUALITY FOR THESIS
    # =========================================================================
    print("\nüîç CHECK 8: Publication-Quality Visualizations")
    print("-" * 80)
    
    figures_created = []
    
    if 'fig_rag_analysis' in globals():
        fig = globals()['fig_rag_analysis']
        dpi = fig.dpi
        width, height = fig.get_size_inches()
        
        print(f"  ‚Ä¢ Figure 3 (RAG Variant Analysis):")
        print(f"    - Size: {width:.1f}\" √ó {height:.1f}\" (thesis recommended: 7-10\")")
        print(f"    - DPI: {dpi} (thesis minimum: 300 dpi)")
        
        if width >= 7 and width <= 12:
            validation_results['checks'].append(('Figure Size', 'Fig 3', True))
        else:
            validation_results['warnings'].append(
                f"Figure 3 size ({width:.1f}\") outside recommended range"
            )
        
        figures_created.append('fig_rag_analysis')
    
    if 'fig_judge_analysis' in globals():
        fig = globals()['fig_judge_analysis']
        dpi = fig.dpi
        width, height = fig.get_size_inches()
        
        print(f"  ‚Ä¢ Figure 4 (Judge-Level Analysis):")
        print(f"    - Size: {width:.1f}\" √ó {height:.1f}\"")
        print(f"    - DPI: {dpi}")
        
        figures_created.append('fig_judge_analysis')
    
    print(f"\n  ‚úì {len(figures_created)} thesis-quality figures generated")
    validation_results['checks'].append(('Visualizations', 'Created', len(figures_created) >= 2))
    
    # =========================================================================
    # SUMMARY REPORT
    # =========================================================================
    print("\n" + "="*80)
    print("VALIDATION SUMMARY")
    print("="*80)
    
    n_checks = len(validation_results['checks'])
    n_passed = sum(1 for _, _, passed in validation_results['checks'] if passed)
    pass_rate = n_passed / n_checks if n_checks > 0 else 0
    
    validation_results['passed_checks'] = n_passed
    validation_results['total_checks'] = n_checks
    validation_results['pass_rate'] = pass_rate
    
    print(f"\nüìä Overall Quality Score: {n_passed}/{n_checks} checks passed ({pass_rate:.1%})")
    
    if len(validation_results['critical_failures']) == 0:
        print("\n‚úÖ ALL CRITICAL CHECKS PASSED")
        print("   Analysis meets PhD thesis publication standards")
    else:
        print(f"\n‚ùå {len(validation_results['critical_failures'])} CRITICAL FAILURES DETECTED:")
        for failure in validation_results['critical_failures']:
            print(f"   ‚Ä¢ {failure}")
    
    if validation_results['warnings']:
        print(f"\n‚ö†Ô∏è  {len(validation_results['warnings'])} WARNINGS (non-critical):")
        for warning in validation_results['warnings'][:5]:  # Show first 5
            print(f"   ‚Ä¢ {warning}")
    
    print("\n" + "="*80)
    print("‚úì VALIDATION COMPLETE")
    print("="*80)
    
    return validation_results


# =============================================================================
# EXECUTE COMPREHENSIVE VALIDATION
# =============================================================================

print("\nExecuting comprehensive PhD thesis quality validation...\n")
phd_validation_report = comprehensive_phd_quality_validation()
```

### 3.X Secondary Exploratory Objective: RAG Chunking Strategy Analysis

**Research Question**: Does retrieval granularity affect inter-rater reliability in LLM-as-a-Judge evaluations?

#### 3.X.1 Data Source and Sampling

The NewHQ organizational policy corpus provided 520 paired evaluations across four RAG chunking strategies: (1) v0: document-level indexing (coarse, N=130), (2) v1: page-level OCR indexing (medium, N=130), (3) v2: page-level layout indexing (medium, N=130), and (4) v3: topic-clustered indexing (fine, N=130). The same 130 questions were evaluated across all variants, creating a balanced within-subjects design. Both judges (GPT-4o-mini and Gemini Flash 2.0) independently evaluated identical system outputs, enabling paired reliability analysis.

#### 3.X.2 Statistical Framework

This exploratory analysis employed descriptive and comparative methods exclusively from the primary reliability study to maintain methodological consistency. No inferential tests were introduced beyond the established framework.

**Reliability Metrics**:
- **Spearman's rank correlation coefficient (œÅ)**: Assesses monotonic association between judges' ordinal ratings (Spearman, 1904). Appropriate for 5-point Likert scale data where interval properties cannot be assumed. Implementation used `scipy.stats.spearmanr` with average rank method for tie handling.
- **Weighted Cohen's kappa (Œ∫)**: Measures inter-rater agreement with quadratic weights penalizing larger disagreements (Cohen, 1968; Fleiss & Cohen, 1973). Weights defined as: w_ij = 1 - ((i-j)¬≤/(k-1)¬≤) where k=5 categories. Implementation used `sklearn.metrics.cohen_kappa_score` with `weights='quadratic'`.

**Interpretation Thresholds**:
- Correlation strength (Evans, 1996): |œÅ| < 0.40 (weak), 0.40-0.59 (moderate), 0.60-0.79 (strong), ‚â•0.80 (very strong)
- Agreement levels (Landis & Koch, 1977): Œ∫ < 0.00 (poor), 0.00-0.20 (slight), 0.21-0.40 (fair), 0.41-0.60 (moderate), 0.61-0.80 (substantial), 0.81-1.00 (almost perfect)

**Stratification Approach**:
Reliability metrics computed separately for each RAG variant √ó evaluation dimension combination (4 variants √ó 5 dimensions = 20 strata). This stratification enabled identification of dimension-specific sensitivity to retrieval granularity.

**Missing Data**: Listwise deletion applied per stratum (complete-case analysis). Out-of-range values (<1 or >5 on Likert scale) set to NA prior to analysis. Overall missing data rate: <5%.

#### 3.X.3 Analytical Plan

1. **Stratified Reliability Analysis**: Computed Spearman œÅ and weighted Œ∫ for each variant-dimension combination
2. **Descriptive Comparison**: Calculated mean reliability across variants with standard deviations
3. **Aggregate Performance Analysis**: Examined mean evaluation scores (1-5 scale) by variant and judge
4. **Judge-Level Pattern Analysis**: Identified individual judge scoring tendencies and variant-specific sensitivities
5. **Visualization**: Generated multi-panel figures showing reliability patterns and judge-level differences

**Methodological Constraints**: 
- Exploratory (hypothesis-generating) rather than confirmatory analysis
- Non-independence acknowledged: same questions across variants limit causal inference
- Results inform future confirmatory research with independent question sets per variant
